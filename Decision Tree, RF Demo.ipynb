# Quiz 2: Ensemble methods

Decision trees, bagging, random forests, boosted trees

### 1. California Housing Dataset

Load the built-in California Housing dataset

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
```

Split data into train/test sets

```python
X, Y = housing.data, housing.target

print(X.shape)
print(Y.shape)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(X_test.shape)
print(y_test.shape)
```

### 2. Decision trees

Build a decision tree regressor on the training set. Make predictions on test set and Evaluate the predictions. Print out error on the test set.

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error


clf = DecisionTreeRegressor(random_state=0)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
score = clf.score(X_test, y_test)

print("MSE : ",mse)
print("Score : ", score)
```

### 3. Random forests


*   Construct Random Forests Regressors with n_estimators (number of trees) being: 2, 3, 5, 10, 20, 50, 100, 200
*   Fit the models on the training set. Compute MSE loss on the test set
*   Produce a plot of #trees vs. test loss





```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
RF_MSE =[]
estimators = [2,3,5,10,20,50,100,200]

for k in estimators:
    rfc = RandomForestRegressor(n_estimators=k)
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    RF_MSE.append(mean_squared_error(y_test, y_pred))

import matplotlib.pyplot as plt

plt.plot(estimators, RF_MSE)
plt.xlabel('Trees')
plt.ylabel('Test Loss')
```
